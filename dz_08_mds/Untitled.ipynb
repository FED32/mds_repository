
Задание 1

Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

%matplotlib inline
```

```python
# Нормальное распределение
```

```python
sigma = 0.5
x = np.linspace(-15,15,100)
for mu in np.linspace(-10, 10, 5):
    norm_rv = sts.norm(loc=mu, scale=sigma)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
mu = 2
x = np.linspace(-5,10,100)
for sigma in np.linspace(1, 5, 5):
    norm_rv = sts.norm(loc=mu, scale=sigma)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
# Экспоненциальное распределение
```

```python
x = np.linspace(-11,11,1000)
for lam in np.linspace(-10, 5, 10):
    expon_rv = sts.expon(lam)
    pdf = expon_rv.pdf(x)
    plt.plot(x, pdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
# распределение Стьюдента
```

```python
x = np.linspace(-5,5,1000)
for n in np.linspace(-10, 10, 10):
    t_rv = sts.t(n)
    pdf = t_rv.pdf(x)
    plt.plot(x, pdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```
