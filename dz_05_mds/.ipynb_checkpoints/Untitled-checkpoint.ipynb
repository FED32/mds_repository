
Задание 1:¶

Посчитайте частные производные функций (как обычно, необходимо решение на бумажке или в LaTeX и на scipy):

1) $f(x,y)=2x^2y^3 + 1/x + y^2x + 7$, а для этой функции также градиент в точке $(1,2)$

2) $f(x,y)=x^2y - sin(xy) + cos(x^2) + 6y$


```python
import numpy as np
from scipy.optimize import approx_fprime
```

```python
def func(x):
    return 2*(x[0]**2)*(x[1]**3) + (1/x[0]) + (x[1]**2)*x[0] + 7
```

```python
x = np.array([1,2])
```

```python
eps = np.sqrt(np.finfo(float).eps)
approx_fprime(x, func, [eps, eps])
```

```python
def func_1(x1):
    return (x1[0]**2)*x1[1] - np.sin(x1[0]*x1[1]) + np.cos(x1[0]**2) + 6*x1[1]
```

```python
x1 = np.ones(2)
```

```python
eps = np.sqrt(np.finfo(float).eps)
approx_fprime(x1, func_1, [eps, eps])
```
