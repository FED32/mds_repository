```python
def num_of_successes(n, k):
    return factorial(n)/(factorial(k) * factorial(n - k))
```

```python
def probability_of_success(p, n, k):
    C_kn = num_of_successes(n, k)
    return C_kn * (p**k) * (1 - p)**(n - k)
```
